```python
import matplotlib.pyplot as plt

s1 = [2,3,4,5,6,7,1,2]
s2 = [9,1,4,3,2,9,6,3]
days = [1,2,3,4,5,6,7,8]

plt.plot(days,s1)
```

```python
plt.title('Study Time Graph')
plt.xlabel("hours")
plt.ylabel("days")
plt.plot(days,s1)
plt.plot(days,s2)
```

```python
import pandas as pd 
df = pd.read_csv('enrollment_data.csv') 
```

```python
df
```

```python
plt.xlabel('Year')
plt.ylabel('Count of enrollment')
plt.title('studnet enrollment over year')
plt.plot(df['Year'],df['Programming'])
```

```python
plt.xlabel('Year')
plt.ylabel('Count of enrollment')
plt.title('studnet enrollment over year')
plt.plot(df['Year'],df['Programming'])
plt.plot(df['Year'],df['Digital Marketing'])
```

```python
plt.xlabel('Year')
plt.ylabel('Count of enrollment')
plt.title('studnet enrollment over year')
plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing')

plt.legend()
plt.grid()
```
