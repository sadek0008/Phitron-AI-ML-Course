## Custom Build Function

```python
def gradeing(marks):
    if marks >=260:
        return 'A+'
    elif marks >=250:
        return 'A'
    elif marks >=240:
        return 'A-'
    else :
        return 'B'

df['Grade'] = df['Total_marks'].apply(gradeing)
```

```python
df
```

```python
def mark(df):
    a = df['Python Marks']*2
    b=df['Data Structure Marks']*2
    c=df['Algorithm Marks']*2

    return a+b+c

df['ecceptional_mark']=df.apply(mark,axis =1)
```
